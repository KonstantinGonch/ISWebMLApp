```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
```

```python
f = open('learnData.json')
data = json.load(f)
```

```python
data
```

```python
for i in data:
    if i['cpu'] > 20 or i['ram'] > 60:
        i['threat'] = 1
    else:
        i['threat'] = 0
```

```python
dat = pd.DataFrame(data)
```

```python
dat['cpu'].hist()
```

```python
x = dat.drop('threat', axis = 1)
X = x.values
y = dat['threat']
Y = y.values
```

```python
knn_model = KNeighborsRegressor(n_neighbors=3)
```

```python
knn_model.fit(X, Y)
```

```python
knn_model.predict([[2,80]])
```
